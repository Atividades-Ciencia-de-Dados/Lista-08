```python
import re
import spacy

# Carregando o modelo de português
nlp = spacy.load("pt_core_news_lg") # nlp = spacy.load("pt_core_news_sm")

# Função para limpar o texto
def processar_texto(texto):
    """
    Limpa o texto removendo emojis e caracteres especiais irrelevantes,
    mantendo acentuação e capitalização (essenciais para NER).
    """
    texto = re.sub(r'[^\w\s,.!?áéíóúãõâêîôûçÁÉÍÓÚÃÕÂÊÎÔÛÇ-]', '', texto)  # remove emojis e símbolos
    texto = texto.strip()
    texto = re.sub(r'\s+', ' ', texto)
    return texto

# Função para extrair entidades úteis
def extrair_entidades(texto):
    doc = nlp(texto)
    return list(set(ent.text.strip() for ent in doc.ents if ent.label_ in ['PER', 'LOC', 'ORG', 'GPE']))

# Aplicar o pré-processamento
mensagens_limpa = [processar_texto(m) for m in mensagens]

# Aplicar extração de entidades
entidades_por_mensagem = [extrair_entidades(m) for m in mensagens_limpa]

# Filtrar mensagens com pelo menos 2 entidades
entidades_por_mensagem = [e for e in entidades_por_mensagem if len(e) >= 2]
```

### Preparar os dados: transformar as listas em transações

```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

# Codificar transações para DataFrame binário
te = TransactionEncoder()
te_ary = te.fit(entidades_por_mensagem).transform(entidades_por_mensagem)
df_transacoes = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Calcular o suporte de cada item (frequência relativa)
item_support = df_transacoes.mean(axis=0)

# Selecionar colunas com suporte >= 0.001
cols_frequentes = item_support[item_support >= 0.001].index

# Criar DataFrame com apenas os itens frequentes
df_filtrado = df_transacoes[cols_frequentes]
```

```python
print("Tamanho df_transações:", df_transacoes.shape)
print("Tamanho df_filtrado:", df_filtrado.shape)
```

### Apriori

Algoritmo que usa dados no formato de listas de transações.
Gera itemsets frequentes em etapas:
- Conta itens individuais (1-itemsets).
- Combina para formar pares (2-itemsets), trios (3-itemsets) e assim por diante.- Só continua com combinações se os subconjuntos forem frequentes.

Usa o princípio:
- **Se um itemset não é frequente, seus supersets também não serão.**

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Aplicar Apriori (suporte mínimo 0.05 = 0.5%)
frequent_itemsets_apriori = apriori(df_filtrado, min_support=0.005, use_colnames=True)

# Gerar regras de associação com confiança mínima de 60%
rules_apriori = association_rules(frequent_itemsets_apriori, metric="confidence", min_threshold=0.6)
```

### ECLAT (Equivalence Class Clustering and bottom-up Lattice Traversal)

- Usa a representação vertical: para cada item, guarda em quais transações ele aparece.
- Gera novos itemsets fazendo interseções entre listas de transações.
- É recursivo e eficiente para conjuntos densos.

```python
from collections import defaultdict
import pandas as pd

def eclat(transactions, min_support=0.005):
    total_trans = len(transactions)
    min_count = int(total_trans * min_support)

    # 1. Construir a tabela vertical: item -> conjunto de TIDs
    item_tidsets = defaultdict(set)
    for tid, transaction in enumerate(transactions):
        for item in set(transaction):
            item_tidsets[frozenset([item])].add(tid)

    # 2. Função recursiva para gerar itemsets frequentes
    def recursive_eclat(prefix, items):
        results = []  # Lista que armazenará os itemsets frequentes com seus suportes
        for i in range(len(items)):
            item_i = items[i]
            tids_i = item_tidsets[item_i]
            new_itemset = prefix.union(item_i)
            support = len(tids_i)
            if support >= min_count: # Se o suporte for suficiente, guarda o itemset
                results.append((set(new_itemset), support))
                suffix = items[i+1:]
                new_items = []
                for item_j in suffix:
                    tids_j = item_tidsets[item_j]
                    intersec = tids_i & tids_j
                    if len(intersec) >= min_count:
                        item_tidsets[item_i.union(item_j)] = intersec
                        new_items.append(item_i.union(item_j))
                results.extend(recursive_eclat(new_itemset, new_items))
        return results

    # 3. Executar o algoritmo
    result = recursive_eclat(set(), list(item_tidsets.keys()))

    # 4. Converter para DataFrame com suporte relativo
    df_result = pd.DataFrame(result, columns=['itemset', 'support_abs'])
    df_result['support'] = df_result['support_abs'] / total_trans

    return df_result[['itemset', 'support']]

# Aplicando ECLAT
eclat_df = eclat(entidades_por_mensagem, min_support=0.005)
eclat_df
```

```python
from mlxtend.frequent_patterns import association_rules

# Ajustar nome da coluna para o formato esperado pelo mlxtend
eclat_df = eclat_df.rename(columns={'itemset': 'itemsets'})

# Gerar regras de associação a partir dos itemsets frequentes do ECLAT
rules_eclat = association_rules(eclat_df, metric="confidence", min_threshold=0.6)

# Mostrar as regras importantes
print("🔹 Regras ECLAT:")
rules_eclat[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```
